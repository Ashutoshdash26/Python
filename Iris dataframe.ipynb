# Exploring Iris Dataset

```python
from sklearn.datasets import load_iris

Iris = load_iris()
```

```python
print("Iris data :- ",Iris.data)
print("\nrows & column :- ",Iris.data.shape)
```

```python
print("Iris.target :- \n",Iris.target)
print("\nsize :- ",Iris.target.size)
```

```python
Iris.feature_names
```

```python
data=pd.DataFrame(data=Iris.data,columns=Iris.feature_names)
data
```

```python
data1=pd.DataFrame(data=Iris.data,columns=Iris.feature_names)
data1['target']=Iris.target
data1
```

```python
print("data1[45:50] :- ",data1[45:55])
print("\ndata1[51,60] :-  ",data1[51:60])
print("\ndata1[95:100] :- ",data1[95:100])
print("\ndata1[101:105] :- ",data1[101:105])
print("\ndata1[145:150] :- ",data1[145:150])
```
