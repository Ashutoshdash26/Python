```python
# ================================
# 1. IMPORT LIBRARIES
# ================================
import pandas as pd              # for handling data (CSV, tables)
import numpy as np               # for numerical operations

from sklearn.model_selection import train_test_split   # to split data
from sklearn.preprocessing import LabelEncoder         # to convert text → numbers
from sklearn.metrics import accuracy_score             # to check accuracy

from xgboost import XGBClassifier                      # ML model
import joblib                                          # to save model


# ================================
# 2. LOAD DATASET
# ================================
df = pd.read_csv("data/dataset.csv")   # read your CSV file

print("Dataset:")
print(df.head())   # show first 5 rows


# ================================
# 3. SEPARATE FEATURES & LABEL
# ================================
X = df.drop("disease", axis=1)   # all columns except disease → input
y = df["disease"]                # disease column → output


# ================================
# 4. ENCODE LABELS
# ================================
le = LabelEncoder()              # create encoder
y = le.fit_transform(y)          # convert disease names → numbers

print("\nEncoded Labels:")
print(y)


# ================================
# 5. SPLIT DATA (NO STRATIFY HERE)
# ================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,       # 20% test data
    random_state=42      # for same result every time
)


# ================================
# 6. CREATE MODEL
# ================================
model = XGBClassifier(
    n_estimators=100,    # number of trees
    max_depth=4          # depth of trees (smaller for small data)
)


# ================================
# 7. TRAIN MODEL
# ================================
model.fit(X_train, y_train)


# ================================
# 8. TEST MODEL
# ================================
preds = model.predict(X_test)           # predict on test data
accuracy = accuracy_score(y_test, preds)

print("\nAccuracy:", accuracy)


# ================================
# 9. SAVE MODEL
# ================================
joblib.dump(model, "model.pkl")         # save model
joblib.dump(le, "encoder.pkl")          # save encoder

print("\nModel Saved!")


# ================================
# 10. PREDICTION FUNCTION
# ================================
def predict_disease(data):
    
    # convert input into array
    input_data = np.array([[
        data["age"],
        data["gender"],
        data["bp"],
        data["sugar"],
        data["fever"],
        data["cough"],
        data["chest_pain"],
        data["fatigue"]
    ]])

    # get probabilities
    probs = model.predict_proba(input_data)[0]

    # get best prediction
    idx = probs.argmax()

    return {
        "disease": le.inverse_transform([idx])[0],
        "confidence": round(float(probs[idx]), 3)
    }


# ================================
# 11. TEST PREDICTION
# ================================
sample = {
    "age": 55,
    "gender": 0,
    "bp": 160,
    "sugar": 220,
    "fever": 0,
    "cough": 0,
    "chest_pain": 1,
    "fatigue": 1
}

print("\nPrediction:")
print(predict_disease(sample))
```
